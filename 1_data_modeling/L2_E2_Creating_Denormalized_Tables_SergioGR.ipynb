## Query with multiple JOINs

Let's say we need to get:  

transaction_id  
customer_name  
cashier_name  
year  
albums_sold  
amount_sold

```python
query = """
SELECT
    t.transaction_id,
    t.customer_name,
    e.employee_name AS cashier_name,
    t.year,
    a.album_name AS album_sold,
    s.amount_spent AS amount_sold
FROM transactions2 t
JOIN employees e
ON t.cashier_id = e.employee_id
JOIN albums_sold a
ON t.transaction_id = a.transaction_id
JOIN sales s
ON t.transaction_id = s.transaction_id
"""

try:
    cur.execute(query)
except psycopg2.Error as e: 
    print("Error: select *")
    print (e)

row = cur.fetchone()
while row:
   print(row)
   row = cur.fetchone()
```

We were able to get the information by performing multiple JOINS. However, we need to remember that JOINS are slow and if we have a read heavy workload that requires low latency queries we want to reduce the number of JOINS. Let's think about denormalizing our normalized tables.

## Denormalizing our tables

With denormalization you want to think about the queries you are running and how to reduce the number of JOINS even if that means duplicating data. For example, we need to run the following queries:

**Query 1** : `select transaction_id, customer_name, amount_spent FROM <min number of tables>`  
It should generate the amount spent on each transaction.  
**Query 2**: `select cashier_name, SUM(amount_spent) FROM <min number of tables> GROUP BY cashier_name`  
It should generate the total sales by cashier.

### Query 1
`select transaction_id, customer_name, amount_spent FROM <min number of tables>`  
We already have all the information in the `transactions` table, except from `amount_spent`. If we add this attribute to the table we don't need to do any JOIN at all.  

Table Name: transactions  
column 0: transaction_id  
column 1: customer_name  
column 2: cashier_id  
column 3: year  
column 4: amount_spent


```python
create_transactions_query = """
CREATE TABLE IF NOT EXISTS transactions (
    transaction_id INT,
    customer_name VARCHAR,
    cashier_id INT,
    year INT,
    amount_spent INT
);
"""

try: 
    cur.execute(create_transactions_query)
except psycopg2.Error as e: 
    print("Error: Issue creating table")
    print (e)
    
insert_transactions_query = """
INSERT INTO transactions (transaction_id, customer_name, cashier_id, year, amount_spent)
VALUES (%s, %s, %s, %s, %s)
"""

try: 
    cur.execute(insert_transactions_query, (1, "Amanda", 1, 2000, 40))
except psycopg2.Error as e: 
    print("Error: Inserting Rows")
    print (e)
    
try: 
    cur.execute(insert_transactions_query, (2, "Toby", 1, 2000, 19))
except psycopg2.Error as e: 
    print("Error: Inserting Rows")
    print (e)
    
try: 
    cur.execute(insert_transactions_query, (3, "Max", 2, 2018, 45))
except psycopg2.Error as e: 
    print("Error: Inserting Rows")
    print (e)
```

```python
query = """
SELECT transaction_id, customer_name, amount_spent
FROM transactions;
"""
try: 
    cur.execute(query)
except psycopg2.Error as e: 
    print("Error: select *")
    print (e)

row = cur.fetchone()
while row:
   print(row)
   row = cur.fetchone()
```

### Query 2
`select cashier_name, SUM(amount_spent) FROM <min number of tables> GROUP BY cashier_name`  
To avoid using any JOINS, we first create a new table with just the information we need. 
 

Table Name: cashier_sales  
column 0: transaction_id    
column 1: cashier_id  
column 2: cashier_name  
column 3: amount_spent

```python
create_cashier_sales_query = """
CREATE TABLE IF NOT EXISTS cashier_sales (
    transaction_id INT,
    cashier_id INT,
    cashier_name VARCHAR,
    amount_spent INT
);
"""

try: 
    cur.execute(create_cashier_sales_query)
except psycopg2.Error as e: 
    print("Error: Issue creating table")
    print (e)
    
insert_cashier_sales_query = """
INSERT INTO cashier_sales (transaction_id, cashier_id, cashier_name, amount_spent)
VALUES (%s, %s, %s, %s)
"""

try: 
    cur.execute(insert_cashier_sales_query, (1, 1, "Sam", 40))
except psycopg2.Error as e: 
    print("Error: Inserting Rows")
    print (e)
    
try: 
    cur.execute(insert_cashier_sales_query, (2, 1, "Sam", 19))
except psycopg2.Error as e: 
    print("Error: Inserting Rows")
    print (e)
    
try: 
    cur.execute(insert_cashier_sales_query, (3, 2, "Bob", 45))
except psycopg2.Error as e: 
    print("Error: Inserting Rows")
    print (e)
```

```python
query = """
SELECT cashier_name, SUM(amount_spent)
FROM cashier_sales
GROUP BY cashier_name;
"""
try: 
    cur.execute(query)
except psycopg2.Error as e: 
    print("Error: select *")
    print (e)

row = cur.fetchone()
while row:
   print(row)
   row = cur.fetchone()
```
